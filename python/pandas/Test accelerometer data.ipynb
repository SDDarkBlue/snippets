# Tracking bikes

**idea:** use existing android apps to track the accelerometer of someone riding a bycicle. Try to correlate that to state of the road.

## Part 1: explore a first track

I tracked my accelerometer while driving over a bad piece of road. I first drove over it with my phone in the pocket of my coat, then turned and drove over it in the opposite direction with my phone in the pocket of my pants.

First, let's read the data

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import time
import numpy as np
infile = "accelerodata.txt"
aggregated_seconds = 1 # data will be aggregated by this number of seconds. The larger this number, the more the data gets smoothed.

def read_acc_data(infile):
    x = []
    y = []
    z = []
    reader = csv.reader(open(infile), delimiter=' ')
    for i in range(8):
        reader.next() # remove metadata
    for row in reader:
        if row[0][0] is not '#':
            x.append(float(row[0]))
            y.append(float(row[1]))
            z.append(float(row[2]))
    return pd.DataFrame(data={'x': x, 'y': y, 'z': z})

df = read_acc_data(infile)
```

The `time` column contains # miliseconds from the last point. I'd rather have an index in seconds since the start of tracking.

```python
def add_time_index(df):
    time_index = pd.Series(range(0, 50 * len(df.x), 50))
    time_index = time_index / 1000.0
    seconds = aggregated_seconds*floor(time_index/aggregated_seconds)
    df['seconds'] = seconds
    df.index = pd.Index(time_index)
    
add_time_index(df)
```

Plot the complete track

```python
df[['x', 'y', 'z']].plot(style='x', alpha=0.3)
```

About halfway (at ~100 seconds) must be the point where I stopped, took out my phone and put it in the pocket of my pants. You see a big spike in the accelerometer data. Then I turn, quickly have to stop to let a car pass (at ~120 seconds) and continue. You see that the data with the phone in my pants is way more noisy than the first part.

The first part, with the phone is my coat, shows what I was hoping for. A very noisy part, followed by a more smooth part. Let's zoom in on that part of the data.

```python
df.loc[:100][['x','y', 'z']].plot(style='x', alpha=0.3)
```

If we could measure the noisyness in the data, we might get a good correlation with the quality of the road. I'll use the variance in the data as a proxy for noisyness. I'll also take the average variance of x, y and z to get to a final 'shakyness' variable. The real beginning and ending of this part of the track will also be clipped, as they correspond to moments where I take out my phone.

```python
grouped = df.groupby('seconds')
g = grouped.aggregate(np.var)
g['shakyness'] = (g['x'] + g['y'] + g['z']) / 3.0
g['shakyness'].loc[28:93].plot()
```
