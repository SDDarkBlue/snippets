```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import time
import numpy as np
infile = "accelerodata.txt"
aggregated_seconds = 2 # data will be aggregated by this number of seconds. The larger this number, the more the data gets smoothed.
x = []
y = []
z = []
reader = csv.reader(open(infile), delimiter=' ')
for i in range(8):
    reader.next() # remove metadata
for row in reader:
    if row[0][0] is not '#':
        x.append(float(row[0]))
        y.append(float(row[1]))
        z.append(float(row[2]))
    
df = pd.DataFrame(data={'x':x, 'y':y, 'z':z})
time_index = pd.Series(range(0, 50 * len(x), 50))
time_index = time_index / 1000.0
seconds = aggregated_seconds*floor(time_index/aggregated_seconds)
df['seconds'] = seconds
df.index = pd.Index(time_index)
```

```python
df.loc[:100][['x','y', 'z']].plot()
```

```python
grouped = df.groupby('seconds')
g = grouped.aggregate(np.var)
g['shakyness'] = (g['x'] + g['y'] + g['z']) / 3.0
g['shakyness'].loc[:100].plot()
```
